```python
# Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Understanding the composition of cab data
cab_data = pd.read_csv('/Users/aravind/Desktop/CS:DS/Internship-Projects/G2M Insights for Cab Investment /DataSets/Cab_Data.csv')
cab_data.drop_duplicates # No duplicates found
cab_data['Profit'] = cab_data['Price Charged'] - cab_data['Cost of Trip']
cab_data.tail()

# Calculating NA values in all columns
#print(cab_data.isna().sum())

# No NA values found
```

```python
# Understanding the composition of city data
city_data = pd.read_csv('/Users/aravind/Desktop/CS:DS/Internship-Projects/G2M Insights for Cab Investment /DataSets/City.csv')
city_data.tail()

# Calculating NA values in all columns
#print(city_data.isna().sum())

# No NA values found
```

```python
# Understanding the composition of the customer data
customer_data = pd.read_csv('/Users/aravind/Desktop/CS:DS/Internship-Projects/G2M Insights for Cab Investment /DataSets/Customer_ID.csv')
customer_data.drop_duplicates # No duplicates found
customer_data.tail()

# Calculating NA values in all columns
#print(customer_data.isna().sum())

# No NA values found
```

```python
# Understanding the composition of Transaction data
transaction_data = pd.read_csv('/Users/aravind/Desktop/CS:DS/Internship-Projects/G2M Insights for Cab Investment /DataSets/Transaction_ID.csv')
transaction_data.drop_duplicates # No duplicates found
transaction_data.tail()

# Calculating NA values in all columns
#print(transaction_data.isna().sum())

# No NA values found
```

```python
# Looking for  outliers
cab_data.describe()

# We can see the distribution of data based on the descriptive statistics, but to verify outliers and relationships, 
# we need to visualize the data
```

```python
# Let's check for outliers on the following columns of data: Price Charged, and Cost of Trip
# Using a box plot - there appear to be no outliers on the cost of the trip, but there are quite a few on the price charged
boxplot_cost_of_trip_and_price_charged = cab_data.boxplot(column=['Price Charged', 'Cost of Trip'])

# Now that we know that the price charged is a column with many outliers in it, we can focus on our analysis towards that column
# And try and see what insights that we can gather from it, while also seeing the visual distribution of cost of trip
```

```python
# Let's visualize the data in accordance with the following question: 
# Did the date of travel have an effect on trip costs?
cab_data.plot.scatter(x='Date of Travel', y='Cost of Trip')
plt.axis([42000, 44000, 0, 750])
# We can see that the cost of the trip stayed about the same throughout the given timeframe of the data in the problem
# During days between 42500, 43000, and 43250 seemed to have the highest fluctuations in the cost of the trip
# Suggestion: We can suggest to G2M that the cost of the trip for the drivers that they will place in their cabs after
# investing the cab industry will stay about constant throughout the december timeframe given in the problem
```

```python
# Let's visualize the data in accordance with the following question: 
# Did the date of travel have an effect on Price Charged?
cab_data.plot.scatter(x='Date of Travel', y='Price Charged')
plt.axis([42000, 44000, 0, 2500])
# We can see that the price charged fluctuated quite a bit from the mean of 423 during the given timeframe in the problem
# During days 42625, 42850, and once again on day 43150, the highest prices were charged
# Suggestion: We can suggest to G2M that the price charged can be either hiked to the prices on days 42625, 42850, and 43150 for a few more days using different rates 
# Or, we can suggest G2M to reduce prices to increase customers paying for cabs, and hence, eventually causing a boost in fares
```

```python
# Let's visualize the data in accordance with the following question: 
# Did the city in which the trip occurred have any effect on the profit for the trip?
plt.figure(figsize=(8,6))
ax = sns.barplot(x=cab_data['City'], y=cab_data['Profit'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

# We can see that the profit in certain cities like New York, Dallas, etc. were much higher than other cities
# like Nashville, and Phoenix
# Suggestion: It is better for GM2 to start their investment in a city like New York, where the profit margins are high
# Which would make the investment project a success
```

```python
# Joining the transaction_data and customer_data csv files on Customer ID
combined_df = pd.merge(customer_data, transaction_data, how='inner')
combined_df.tail()
```

```python
# Joining the combined_df with the cab_data dataframe on transaction ID
final_combined_df = pd.merge(combined_df, cab_data, how='inner')
final_combined_df.tail()
```

```python
# Let's visualize the data in accordance with the following question: 
# Did the gender of the customer have an effect on the profit for the trip?
plt.figure(figsize=(8,6))
ax = sns.barplot(x=final_combined_df['Gender'], y=final_combined_df['Profit'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

# We can see that the profit was more from the Male gender, compared to the female gender as Male gender customer's seem
# to be paying more on average at around 140$, compared to female customers who give cabs a smaller profit of 125-130$
# Suggestion: G2M should find cities where the population of males is higher to start off with higher sales, as can be
# seen from the data that male customers on average pay more to cabs, causing an increase in profit
```

```python
# Let's visualize the data in accordance with the following question: 
# Did the age of the customer have an effect on the price charged from the trip?
plt.figure(figsize=(8,6))
ax = sns.barplot(x=final_combined_df['Age'], y=final_combined_df['Profit'])
ax.set_xticklabels(ax.get_xticklabels(), rotation=0, ha="right")
plt.tight_layout()
plt.show()

# We can see that the ages of customers who cause the highest profit in cab fares are within the late 20s to mid 50s
# Suggestion: 
```
